```python
import numpy as np
import skfuzzy as fuzz
import skfuzzy.control as ctrl
import matplotlib.pyplot as plt

food_type = ctrl.Antecedent(np.arange(0, 3, 1), 'food_type')
food_weight = ctrl.Antecedent(np.arange(0, 3, 1), 'food_weight')
initial_temp = ctrl.Antecedent(np.arange(0, 3, 1), 'initial_temp')
water_content = ctrl.Antecedent(np.arange(0, 3, 1), 'water_content')

cooking_time = ctrl.Consequent(np.arange(0, 3, 1), 'cooking_time')
heat_level = ctrl.Consequent(np.arange(0, 3, 1), 'heat_level')

labels = ['low', 'medium', 'high']
food_type.automf(names=['soft', 'medium', 'hard'])
food_weight.automf(names=['light', 'medium', 'heavy'])
initial_temp.automf(names=['cold', 'warm', 'hot'])
water_content.automf(names=['low', 'medium', 'high'])
cooking_time.automf(names=['short', 'medium', 'long'])
heat_level.automf(names=['low', 'medium', 'high'])

rules = [
    ctrl.Rule(food_weight['light'] & food_type['soft'], cooking_time['short']),
    ctrl.Rule(food_weight['medium'] & food_type['medium'], cooking_time['medium']),
    ctrl.Rule(food_weight['heavy'] & food_type['hard'], cooking_time['long']),
    ctrl.Rule(initial_temp['cold'], cooking_time['long']),
    ctrl.Rule(initial_temp['warm'], cooking_time['medium']),
    ctrl.Rule(initial_temp['hot'], cooking_time['short']),
    ctrl.Rule(water_content['low'], heat_level['low']),
    ctrl.Rule(water_content['medium'], heat_level['medium']),
    ctrl.Rule(water_content['high'], heat_level['high']),
    ctrl.Rule(food_type['hard'] & food_weight['heavy'] & initial_temp['cold'], cooking_time['long']),
    ctrl.Rule(food_type['hard'] & food_weight['heavy'] & initial_temp['cold'], heat_level['high']),
    ctrl.Rule(food_type['medium'] & food_weight['medium'] & initial_temp['warm'], cooking_time['medium']),
    ctrl.Rule(food_type['medium'] & food_weight['medium'] & initial_temp['warm'], heat_level['medium']),
    ctrl.Rule(food_type['soft'] & food_weight['light'] & initial_temp['hot'], cooking_time['short']),
    ctrl.Rule(food_type['soft'] & food_weight['light'] & initial_temp['hot'], heat_level['low'])
]

cooking_ctrl = ctrl.ControlSystem(rules)
cooking_sim = ctrl.ControlSystemSimulation(cooking_ctrl)

cooking_sim.input['food_type'] = 2
cooking_sim.input['food_weight'] = 2
cooking_sim.input['initial_temp'] = 0
cooking_sim.input['water_content'] = 1

cooking_sim.compute()

print(f"Thời gian nấu: {cooking_sim.output['cooking_time']:.2f} (0: ngắn, 1: trung bình, 2: dài)")
print(f"Mức nhiệt: {cooking_sim.output['heat_level']:.2f} (0: thấp, 1: trung bình, 2: cao)")

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 7))

food_type.view(sim=cooking_sim, ax=axes[0, 0])
food_weight.view(sim=cooking_sim, ax=axes[0, 1])
initial_temp.view(sim=cooking_sim, ax=axes[1, 0])
water_content.view(sim=cooking_sim, ax=axes[1, 1])

plt.tight_layout()
plt.show()

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
cooking_time.view(sim=cooking_sim, ax=axes[0])
heat_level.view(sim=cooking_sim, ax=axes[1])

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt

order_density = ctrl.Antecedent(np.arange(0, 11, 1), 'order_density')
delivery_urgency = ctrl.Antecedent(np.arange(0, 11, 1), 'delivery_urgency')
driver_load = ctrl.Antecedent(np.arange(0, 11, 1), 'driver_load')
traffic_conditions = ctrl.Antecedent(np.arange(0, 11, 1), 'traffic_conditions')
profit_per_delivery = ctrl.Antecedent(np.arange(0, 11, 1), 'profit_per_delivery')

orders_to_combine = ctrl.Consequent(np.arange(0, 11, 1), 'orders_to_combine')
delivery_priority = ctrl.Consequent(np.arange(0, 11, 1), 'delivery_priority')

order_density['low'] = fuzz.trimf(order_density.universe, [0, 0, 5])
order_density['medium'] = fuzz.trimf(order_density.universe, [2, 5, 8])
order_density['high'] = fuzz.trimf(order_density.universe, [5, 10, 10])

delivery_urgency['low'] = fuzz.trimf(delivery_urgency.universe, [0, 0, 5])
delivery_urgency['medium'] = fuzz.trimf(delivery_urgency.universe, [2, 5, 8])
delivery_urgency['high'] = fuzz.trimf(delivery_urgency.universe, [5, 10, 10])

driver_load['low'] = fuzz.trimf(driver_load.universe, [0, 0, 5])
driver_load['medium'] = fuzz.trimf(driver_load.universe, [2, 5, 8])
driver_load['high'] = fuzz.trimf(driver_load.universe, [5, 10, 10])

traffic_conditions['low'] = fuzz.trimf(traffic_conditions.universe, [0, 0, 5])
traffic_conditions['medium'] = fuzz.trimf(traffic_conditions.universe, [2, 5, 8])
traffic_conditions['high'] = fuzz.trimf(traffic_conditions.universe, [5, 10, 10])

profit_per_delivery['low'] = fuzz.trimf(profit_per_delivery.universe, [0, 0, 5])
profit_per_delivery['medium'] = fuzz.trimf(profit_per_delivery.universe, [2, 5, 8])
profit_per_delivery['high'] = fuzz.trimf(profit_per_delivery.universe, [5, 10, 10])

orders_to_combine['few'] = fuzz.trimf(orders_to_combine.universe, [0, 0, 5])
orders_to_combine['some'] = fuzz.trimf(orders_to_combine.universe, [2, 5, 8])
orders_to_combine['many'] = fuzz.trimf(orders_to_combine.universe, [5, 10, 10])

delivery_priority['low'] = fuzz.trimf(delivery_priority.universe, [0, 0, 5])
delivery_priority['medium'] = fuzz.trimf(delivery_priority.universe, [2, 5, 8])
delivery_priority['high'] = fuzz.trimf(delivery_priority.universe, [5, 10, 10])

rule1 = ctrl.Rule(order_density['high'] & driver_load['low'] & traffic_conditions['low'], orders_to_combine['many'])
rule2 = ctrl.Rule(order_density['medium'] & traffic_conditions['high'] & delivery_urgency['medium'], orders_to_combine['some'])
rule3 = ctrl.Rule(driver_load['high'] & order_density['high'] & profit_per_delivery['medium'], orders_to_combine['some'])
rule4 = ctrl.Rule(order_density['low'] & delivery_urgency['high'] & traffic_conditions['medium'], orders_to_combine['some'])
rule5 = ctrl.Rule(profit_per_delivery['high'] & delivery_urgency['high'] & traffic_conditions['high'], orders_to_combine['some'])

rule6 = ctrl.Rule(delivery_urgency['high'] & profit_per_delivery['high'], delivery_priority['high'])
rule7 = ctrl.Rule(delivery_urgency['medium'] & traffic_conditions['medium'], delivery_priority['medium'])
rule8 = ctrl.Rule(delivery_urgency['low'] & order_density['high'] & profit_per_delivery['low'], delivery_priority['low'])

orders_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5])
priority_ctrl = ctrl.ControlSystem([rule6, rule7, rule8])

orders_sim = ctrl.ControlSystemSimulation(orders_ctrl)
priority_sim = ctrl.ControlSystemSimulation(priority_ctrl)

orders_sim.input['order_density'] = 10
orders_sim.input['delivery_urgency'] = 5
orders_sim.input['driver_load'] = 2
orders_sim.input['traffic_conditions'] = 5
orders_sim.input['profit_per_delivery'] = 5

priority_sim.input['delivery_urgency'] = 5
priority_sim.input['profit_per_delivery'] = 5

orders_sim.compute()
priority_sim.compute()

orders_output = orders_sim.output.get('orders_to_combine', 'N/A')
priority_output = priority_sim.output.get('delivery_priority', 'N/A')

print(f"Number of orders to combine: {orders_output}")
print(f"Delivery priority: {priority_output}")

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
order_density.view(ax=axes[0, 0])
delivery_urgency.view(ax=axes[0, 1])
driver_load.view(ax=axes[0, 2])
traffic_conditions.view(ax=axes[1, 0])
profit_per_delivery.view(ax=axes[1, 1])
orders_to_combine.view(sim=orders_sim)
delivery_priority.view(sim=priority_sim)
plt.show()
```
